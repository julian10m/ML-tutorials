```python
import math
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Constants
X_MIN = -10
X_MAX = 10
N_SAMPLES = 50
MU = 0
SIGMA = 1
N_DATASETS = 15

def linear_function(w, b, x_min, x_max):
    if x_min > x_max:
        raise ValueError("x_min should be less than or equal to x_max")
    return [(x, w * x + b) for x in range(x_min, x_max + 1)]

def generate_noisy_samples(w, b, x_min, x_max, n_samples, mu=0, sigma=1):
    x_vals = np.random.uniform(x_min, x_max, n_samples)
    noise = np.random.normal(mu, sigma, n_samples)
    y_vals = w * x_vals + b + noise
    return list(zip(x_vals, y_vals))

def create_datasets(parameters, x_min, x_max, n_samples, mu, sigma):
    datasets = []
    for w, b in parameters:
        line = linear_function(w, b, x_min, x_max)
        samples = generate_noisy_samples(w, b, x_min, x_max, n_samples, mu, sigma)
        datasets.append({"w": w, "b": b, "line": line, "samples": samples})
    return datasets


def plot_datasets(datasets, predictions=None):
    num_plots = len(datasets)
    n_cols = 4
    n_rows = math.ceil(num_plots / n_cols)
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(6 * n_cols, 6 * n_rows))
    axes = axes.flatten()

    for i, data in enumerate(datasets):
        ax = axes[i]
        x_vals, y_vals = zip(*data["line"])
        x_noisy, y_noisy = zip(*data["samples"])

        ax.plot(x_vals, y_vals, 'k--', linewidth=3, label=f"True, unknown - y = {data['w']}x + {data['b']}")
        ax.scatter(x_noisy, y_noisy, color='b', s=100, alpha=0.5, label="Samples")

        if predictions is not None:
            pred_x, pred_y = predictions[i]
            ax.plot(pred_x, pred_y, 'r', linewidth=3, label="Predicted")

        ax.set_title(f"Linear Function: y = {data['w']}x + {data['b']}")
        ax.set_xlabel("x")
        ax.set_ylabel("y")
        ax.axhline(0, color='gray', linewidth=0.5)
        ax.axvline(0, color='gray', linewidth=0.5)
        ax.grid(True)
        ax.legend()

    # Hide any unused subplots
    for j in range(len(datasets), len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()

# Main Execution
def run_linear_regression(datasets):
    model = LinearRegression()
    predictions = []

    for data in datasets:
        x_noisy, y_noisy = zip(*data["samples"])
        X = np.array(x_noisy).reshape(-1, 1)
        y = np.array(y_noisy)
        model.fit(X, y)

        pred_x = np.linspace(X_MIN, X_MAX, 100)
        pred_y = model.predict(pred_x.reshape(-1, 1))
        predictions.append((pred_x, pred_y))

    return predictions
    
def plot_reruns(base_dataset, predictions):
    plt.figure(figsize=(8, 6))
    for pred_x, pred_y in predictions:
        plt.plot(pred_x, pred_y, 'r', linewidth=1, alpha=0.4)
    x_true, y_true = zip(*base_dataset['line'])
    plt.plot(x_true, y_true, 'k--', linewidth=3, label=f"True, unknown - y = {true_w}x + {true_b}")

    plt.title("Fit Variability Across Multiple Re-samplings")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    plt.legend()
    plt.show()    
```

```python
PARAMETERS = [(2, 1)]
datasets = create_datasets(PARAMETERS, X_MIN, X_MAX, N_SAMPLES, MU, SIGMA)
plot_datasets(datasets)
```

```python
predictions = run_linear_regression(datasets)
plot_datasets(datasets, predictions)
```

```python
N_DATASETS = 15
base_dataset = datasets[0]

w, b = base_dataset['w'], base_dataset['b']
datasets = create_datasets([[w, b]] * N_DATASETS, X_MIN, X_MAX, N_SAMPLES, MU, SIGMA)
predictions = run_linear_regression(datasets)

plot_reruns(base_dataset, predictions)
```

```python
PARAMETERS = [(0.5, 1.5), (1.5, -0.5), (-1, 5), (-0.5, -3)]
datasets = create_datasets(PARAMETERS, X_MIN, X_MAX, N_SAMPLES, MU, SIGMA)
predictions = run_linear_regression(datasets)
plot_datasets(datasets, predictions)
```
